Final function to apply our recommender system.

```python
def collaborative_filtering(x):
    fulldata = x.merge(articles, on = 'contentId' )
    artrated_byperson = fulldata.groupby('personId').count()
    # We create a subset of the dataframe containing only users who have evaluated at least 20 articles. This will be the 
    # moment where the collaborative filtering algorithm will start to be applied. 
    artrated_byperson20 = artrated_byperson[artrated_byperson['rating'] > 20]
    # We include eventType to keep the object as a Dataframe rather than as an array. 
    artrated_byperson20 = artrated_byperson20[['rating','eventType']]
    artrated_byperson20 = artrated_byperson20.reset_index()
    artrated_byperson20 = artrated_byperson20[['personId','rating']]
    rt = artrated_byperson20.merge(fulldata, on='personId')
    user_item = rt[['title','personId','rating_y']]
    # Pivot table to locate the title of the articles as features of the dataframe and the ratings as its values.
    user_item = user_item.pivot_table(index = 'personId', columns = 'title', values = 'rating_y')
    useritem_norm = user_item.sub(user_item.mean(axis = 1), axis = 'rows')
    user_similarity = useritem_norm.T.corr()
    # We drop a particular observation to compute its similarity with the rest of users.
    user = 9210530975708218054
    # We store it in a different variable to maintain the original datatset unaffected
    user_similarity2 = user_similarity.drop(index = user)
    n = 10 # number of users with similarity taken into account 
    similarity_threshold = 0.3
    similar_users = user_similarity2[user_similarity2[user] > similarity_threshold][user].sort_values(ascending = False)[:n]
    picked_userid_read = user_item[user_item.index == user].dropna(axis=1, how = 'all')
    similar_user_articles = useritem_norm[useritem_norm.index.isin(similar_users.index)].dropna(axis=1, how='all')
    similar_user_articles.drop(picked_userid_read.columns,axis=1, inplace=True, errors='ignore')
    item_score = {}
    # Loop through items
    for i in similar_user_articles.columns:
      # Get the ratings for article i
      article_rating = similar_user_articles[i]
      # Create a variable to store the score
      total = 0
      # Create a variable to store the number of scores
      count = 0
      # Loop through similar users
      for u in similar_users.index:
        # If the article has rating
        if pd.isna(article_rating[u]) == False:
          # Score is the sum of user similarity score multiply by the article rating
          score = similar_users[u] * article_rating[u]
          # Add the score to the total score for the article so far
          total += score
          # Add 1 to the count
          count +=1
      # Get the average score for the item
      item_score[i] = total / count
    # Convert dictionary to pandas dataframe
    item_score = pd.DataFrame(item_score.items(), columns=['article', 'article_score'])
    
    # Sort the article by score
    ranked_item_score = item_score.sort_values(by='article_score', ascending=False)

    finaldf = pd.merge(ranked_item_score, articles[['title','url']], left_on = 'article', right_on = 'title', how = 'left')
    finaldf = finaldf.drop('title', axis = 1)

    # Average rating for the picked user
    avg_rating = user_item[user_item.index == user].T.mean()[user]
    # Print the average movie rating for user n
    print(f'The average movie rating for user {user} is {avg_rating:.2f}')
    # Calcuate the predicted rating
    finaldf['article_rating'] = finaldf['article_score'] + avg_rating
    # Select top m article
    m = 10
    # Take a look at the data

    return(finaldf.head(m))
```
